```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import ast

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('./csvfile/Scraping_preprocessing_v2.csv')

# Concatenate all the content strings into a single string
df['content']= [ast.literal_eval(item) for item in df['content']]
df['content'] = [' '.join(content) for content in df['content']]

comment_words = ""

df.to_csv(r'Scraping_preprocessing_v3.csv', index = False)

for val in df['content']: 
    val = str(val)
    tokens = val.split()

    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    comment_words += " ".join(tokens)+" "

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(comment_words)

# # Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
