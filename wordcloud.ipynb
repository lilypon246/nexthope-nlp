```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('./datasets/Scraping_preprocessing_v2.csv')

#Concatenate all the content strings into a single string
df['content'] = df['content'].apply(lambda x: [item.strip("' ") for item in ast.literal_eval(x)])
df['content'] = [' '.join(content) for content in df['content']]

comment_words = ""

for val in df['content']: 
    val = str(val)
    tokens = val.split()

    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()

    comment_words += " ".join(tokens)+" "

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(comment_words)

# # Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
